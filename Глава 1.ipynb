1. Какие числовые признаки называются дискретными, а какие — непрерывными? Приведите собственные примеры.

•	Дискретные признаки:
Это признаки, которые принимают конечное или счетное число значений. Обычно они представляют собой целые числа.
Примеры:


*   Количество комнат в квартире (1, 2, 3, ...).
*   Количество студентов в аудитории.
*   Число детей в семье.





•	Непрерывные признаки:

Это признаки, которые могут принимать любое значение в некотором диапазоне. Они обычно представляют собой вещественные числа.
Примеры:

o	Рост человека (например, 175.5 см).

o	Температура воздуха (например, 23.7 °C).

o	Время выполнения задачи (например, 12.345 секунд).

2. Опишите разницу между обучением с учителем и обучением без учителя.

•	Обучение с учителем (Supervised Learning):
Модель обучается на данных, для которых известны правильные ответы (метки). Цель — научиться предсказывать метки для новых данных.

Примеры задач:

*   Классификация (например, определение спама в письмах).
*   Регрессия (например, предсказание цены дома).
*   Обучение без учителя (Unsupervised Learning):

Модель обучается на данных без меток. Цель — найти скрытые структуры или закономерности в данных.

Примеры задач:
*   Кластеризация (например, группировка клиентов по поведению).
*   Понижение размерности (например, PCA для визуализации данных).

3. Опишите разницу между задачами классификации и регрессии.

•	Классификация:

Цель — предсказать категориальную метку (класс) для объекта.
Пример: Определить, является ли изображение кошкой или собакой.

•	Регрессия:

Цель — предсказать непрерывное числовое значение.
Пример: Предсказать стоимость дома на основе его характеристик.

5. Почему в качестве альтернативы простому использованию всех тренировочных данных рекомендуется выполнять проверку модели на валидационных данных?

Использование валидационных данных позволяет оценить, насколько хорошо модель обобщает данные, которые она не видела во время обучения. Это помогает:

•	Избежать переобучения (когда модель слишком хорошо запоминает тренировочные данные, но плохо работает на новых данных).

•	Настроить гиперпараметры модели.

•	Оценить реальную производительность модели перед тестированием на тестовых данных.

6. Даны три матрицы A, B, C. Матрица A имеет размеры 5 × 4, B — размеры 4 × 6, C имеет размеры 3 × 5. Укажите все возможные матрицы, которые можно перемножить между собой.

Для перемножения двух матриц количество столбцов первой матрицы должно совпадать с количеством строк второй матрицы.
•	A (5 × 4) и B (4 × 6): Можно перемножить, результат будет размером 5 × 6.
•	C (3 × 5) и A (5 × 4): Можно перемножить, результат будет размером 3 × 4.

Остальные комбинации невозможны:

•	A (5 × 4) и C (3 × 5): Нельзя, так как 4 ≠ 3.
•	B (4 × 6) и A (5 × 4): Нельзя, так как 6 ≠ 5.
•	B (4 × 6) и C (3 × 5): Нельзя, так как 6 ≠ 3.
•	C (3 × 5) и B (4 × 6): Нельзя, так как 5 ≠ 4.

7. Найдите градиент функции f(x, y, z) = yx² + ln(y) + eᶻ.

Градиент функции — это вектор частных производных по каждой переменной:
∇f = (∂f/∂x, ∂f/∂y, ∂f/∂z).

Вычислим частные производные:

•	По x: ∂f/∂x = 2yx.

•	По y: ∂f/∂y = x² + 1/y.

•	По z: ∂f/∂z = eᶻ.

Итоговый градиент:
∇f = (2yx, x² + 1/y, eᶻ).

8. Найдите частные производные сложной функции E = (y_pred - y)², где y_pred = wx + b, по переменным w и b.

Функция ошибки:
E = (y_pred - y)² = (wx + b - y)².

Найдем частные производные:

•	По w: ∂E/∂w = 2(wx + b - y) · x.

•	По b: ∂E/∂b = 2(wx + b - y).
